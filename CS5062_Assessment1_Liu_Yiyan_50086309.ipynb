```python
x_df = df.drop(['Percent body fat'], axis=1)
y_df = df['Percent body fat'].to_frame()
feature_name = x_df.columns.values
```

```python
from sklearn.preprocessing import StandardScaler
transfer = StandardScaler() 
x_df = transfer.fit_transform(x_df)
x_df = pd.DataFrame(x_df, columns = feature_name )
feature_name
```

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error


def train_cv(X,y,n_splits,alpha,shuffle=True,random_state=30):
    k_fold = KFold(n_splits = n_splits, shuffle=shuffle,random_state=random_state)
    MSE = []
   
    for train_index, test_index in k_fold.split(X):
        X_train, X_test = X.loc[train_index], X.loc[test_index]
        y_train, y_test = y.loc[train_index], y.loc[test_index]
        lasso = Lasso(alpha = alpha)
        lasso.fit(X_train, y_train)
        y_test_pred = lasso.predict(X_test)
        MSE.append(mean_squared_error(y_test, y_test_pred))
    return MSE

# res = train_cv(x_df,y_df,n_splits = 10,alpha=0.01)
# res
```

```python
lambda_lasso = np.arange(0.01,1.01,0.01) # 0.01 is task required
def model_selection_cv(X, Y):
    test_MSE = []
    for alpha in lambda_lasso:
        mse = train_cv(X, Y, n_splits = 10, alpha=alpha)
        test_MSE.append(mse)
    return test_MSE

test_MSE = model_selection_cv(x_df,y_df)
# print(list(test_MSE))
sum_mse = []
for i in test_MSE:
    sum_mse.append(sum(i))
plt.figure(figsize=(6,3))
plt.plot(lambda_lasso, sum_mse)
plt.xlabel('range of lambda')
plt.ylabel('mean square error')

min_index = sum_mse.index(min(sum_mse))#the index of best lambda
best_lambda = lambda_lasso[min_index]#best lambda
best_lambda
print("best lambda:","%.2f" % best_lambda)
```

#### Train the best lambda model

```python
lasso = Lasso(alpha = best_lambda)
lasso.fit(x_df, y_df)
coef = pd.DataFrame(list(zip(feature_name,lasso.coef_)))
coef
```

```python
plt.figure(figsize=(20, 8))
plt.plot(coef.iloc[:,0], coef.iloc[:,1])
plt.grid()
```
